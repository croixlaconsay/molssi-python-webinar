## Plotting and Data Visualization

```python
import os
import numpy

#prepare data for plotting

distance_file = os.path.join('data', 'distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')

headers = distances[0]
data = distances[1:]
data = data.astype(numpy.float)

print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
sample_name = headers[1]

plt.figure()
plt.xlabel('Simulation Frame') # label x-axis
plt.ylabel('Distance (angstrom)') # label y-axis
# to add a name to the plot, just add the title after 'label'
plt.plot(data[:,1], label=sample_name) 
plt.legend()

# What if you want to save this plot as a figure? use 'savefig'

plt.savefig(F'{sample_name}.png', dpi=300) # dpi here sets the resolution
```

# Plotting Multiple Sets of Data

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:,1], label=headers[1])
plt.plot(data[:,2], label=headers[2]) # Just add another line with data 2
plt.legend()
plt.savefig('two_samples.png')
```

```python
plt.figure()

for col in range(1, len(headers)):
    plt.plot(data[:, col], label=headers[col])
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.savefig('all_samples.png')
plt.legend()
```

# Make a different plot for each column of data

```python
#plt.figure()

for col in range(1, len(headers)):
    plt.plot(data[:, col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png')
    plt.figure()

```
